```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a = pd.read_csv ('needfinding.csv')
```

```python
a.head()
```

```python
a.info()
```

```python
b=a['Q1'].value_counts()
b=b/sum(b)
b
c = plt.pie(b, labels=b.index,autopct='%1.1f%%');
plt.title('Age distribution?')
plt.show()
```

```python
b=a['Q2'].value_counts()
b=b/sum(b)
b
c = plt.pie(b, labels=b.index,autopct='%1.1f%%');
plt.title('What is your gender?')
plt.show()
```

```python
answers = []
for i in a['Q3']: 
    elements = i.lower().replace('; ', ', ').replace('/', ', ').replace(' and ', ', ').split (",")
    for j in elements:
        answers.append(j.strip())
answers=pd.Series(answers)
b=answers.value_counts()
b=b/sum(b)
b
c = plt.pie(b, labels=b.index,autopct='%1.1f%%');
plt.title('What is your current job?')
plt.show()

```

```python
b=a['Q4'].value_counts()
b=b/sum(b)
b
c = plt.pie(b, labels=b.index,autopct='%1.1f%%');
plt.title('how many exams taking interfaces you have encountered before?')
plt.show()
```

```python
answers = []
for i in a['Q5']: 
    elements = i.split (";")
    for j in elements:
        answers.append(j)
answers=pd.Series(answers)
b=answers.value_counts()
b=b/sum(b)
b
c = plt.pie(b, labels=b.index,autopct='%1.1f%%');
plt.title('In this interface, can you locate Q&A within 10 seconds?')
plt.show()
```

```python
answers = []
for i in a['Q7']: 
    elements = str(i).split (";")
    for j in elements:
        answers.append(j)
answers=pd.Series(answers)
b=answers.value_counts()
b=b/sum(b)
b
c = plt.pie(b, labels=b.index,autopct='%1.1f%%');
plt.title('which kind of exam-taking interface do you prefer?')
plt.show()
```
