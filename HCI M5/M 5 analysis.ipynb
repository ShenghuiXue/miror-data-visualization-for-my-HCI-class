```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a = pd.read_csv ('survey2.csv')
```

```python
a.info()
```

```python
b=a['Q1'].value_counts()
b=b/sum(b)
b
c = plt.pie(b, labels=b.index,autopct='%1.1f%%');
plt.title('Have you used Google glass or similar products before?')
plt.show()
```
