```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a = pd.read_csv ('nf3.csv')
```

```python
a.info()
```

```python
b=a['Q1'].value_counts()
b=b/sum(b)
b
c = plt.pie(b, labels=b.index,autopct='%1.1f%%');
plt.title('Age distribution')
plt.show()
```

```python
b=a['Q2'].value_counts()
b=b/sum(b)
b
c = plt.pie(b, labels=b.index,autopct='%1.1f%%');
plt.title('Have you taken Udemy class before?')
plt.show()
```

```python
b=a['Q5'].value_counts()
b=b/sum(b)
print (b.index)
index= []
disc = ['', "strongly Disagree", 'Disagree','Neutral','Agree', 'Strongly Agree']
for i in b.index:
     index.append (disc[i])
c = plt.pie(b, labels=index,autopct='%1.1f%%');
plt.title('Rating information of each video [up vote (like) and down vote (dislike) button] will be useful.')
plt.show()
```
